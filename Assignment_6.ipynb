Q1. Ask the user to enter a value (e.g., M) for a mathematical function.

Generate x values from -10 to 10 using np.linspace().

Compute y values for:

* y= M⋅x^2

* y=M⋅sin(x)

Plot both functions in a single figure:

* Use different colors and line styles.

* Add legend, grid, and title.

```python
from pickle import TRUE
import matplotlib.pyplot as plt
import numpy as np
import math
M = float(input("Give a value for M "))
x = np.linspace(-10,10,400)
y1 = M*x**2
y2 = M*np.sin(x)
plt.plot(x,y1,label = "y = M*x^2")
plt.plot(x,y2,label = "y = M*sin(x)")
plt.grid(visible = TRUE)
plt.title("Graph of y = M*x^2 and y = sin(x)")
plt.legend()
plt.show()
```

Q.2 Create a dataset of five subjects and scores.

Convert it into a Pandas DataFrame.

Plot the scores using a Seaborn bar plot with:
* Different colors for each bar.
* Annotations on top of each bar.
* Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sb
dataset = {'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
           'Score': [85, 90, 78, 88, 92]}
df = pd.DataFrame(dataset)
print(df)

sb.set_theme()
sb.barplot(data = df,x = "Subject", y = "Score", hue = "Subject")
plt.title("Barplot of Student Data")
plt.grid(visible=TRUE)
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().

Create a 2x2 subplot layout displaying:

* Line plot showing cumulative sum.
* Scatter plot with random noise.

Customize with titles, axis labels, and grids.

```python
roll_no = 102317004
np.random.seed(roll_no)
x = np.random.randn(50)
fig, plots = plt.subplots(1,2,figsize = (12,6))
plots[0].plot(np.cumsum(x),label = "Cumulative Sum",color = "b")
plots[0].set_xlabel("Number of students")
plots[0].set_ylabel("Sum of Roll Numbers")
plots[0].set_title("Cumulative Sum")
plots[0].grid(visible = True)

plots[1].plot(x)
plots[1].set_title("Scatter Plot with Random Noise")
plots[1].set_xlabel("Roll Numbers")
plots[1].set_ylabel("Randomness")
```
